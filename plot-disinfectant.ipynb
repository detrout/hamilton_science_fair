```python
data = pandas.DataFrame({
        "a": [1.1, 2, 3.1, 4, 5.2, 6.1, 7.3], 
        "b": [0.9, 2.1, 3.2, 4.3, 4.8, 5.8, 6.5],
        "c": [0.95, 2.05, 2.7, 3.7, 5.5, 6.3, 7.5],
    })
data    
```

```python
seaborn.boxplot(data)
```

```python
data = pandas.DataFrame({
        "control disk": [0,0,0], 
        "10% clorox": [32,30,30],
        "95% ethanol": [12,10,10],
        "3% hydrogen peroxide": [66,66,66],
        "100% isoproponol": [12,12,14],
        "vinegar": [20,20,20],
        "windex vinegar": [11,12,10],
        "windex": [10,14,12],
    })
data    
```

```python
#matplotlib.style.use("ggplot")
seaborn.boxplot(data, orient="h")
```

```python
matplotlib.style.use("seaborn-v0_8-darkgrid")
ax= seaborn.boxplot(data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
ax.set_ylabel("exclusion diameter in milimeters")
pass
```

```python
matplotlib.style.use("seaborn-v0_8-pastel")
ax= seaborn.boxplot(data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_ylabel("exclusion diameter in milimeters")
pass
```

```python
matplotlib.style.use("ggplot")
ax= seaborn.boxplot(data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_ylabel("exclusion diameter in milimeters")
pass
```

```python
matplotlib.style.use("seaborn-v0_8-whitegrid")
ax= seaborn.boxplot(data)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_ylabel("exclusion diameter in milimeters")
pass
```
