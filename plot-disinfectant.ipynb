Here I put the measured data into a a table with labels, and have it display the table.

```python
data = pandas.DataFrame({
        "Control disk": [0,0,0], 
        "10% Clorox": [32,30,30],
        "95% Ethanol": [12,10,10],
        "3% Hydrogen peroxide": [66,66,66],
        "100% Isopropanol": [12,12,14],
        "Vinegar": [20,20,20],
        "Windex with Vinegar": [11,12,10],
        "Windex": [10,14,12],
        "Water": [0, 0, 0],
    })
data    
```

This is the figure with the labels on the vertical (y-axis) so they're easy to read, and it's using the "ggplot" style.

```python
matplotlib.style.use("ggplot")
ax = seaborn.boxplot(data, orient="h")
```

This version uses the "seaborn-v0_8-darkgrid" style, has the labels on the bottom, and rotates the 45 degrees so they're kind of readable. see the `rotation=45` argument in `set_xticklabels`

```python
matplotlib.style.use("seaborn-v0_8-darkgrid")
ax = seaborn.boxplot(data)
_ = ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
_ = ax.set_ylabel("exclusion diameter in milimeters")
```

And this version is using the `seaborn-v0_8-pastel` color scheme, with the bottom axis labels rotated 90 degrees.

```python
matplotlib.style.use("seaborn-v0_8-pastel")
ax= seaborn.boxplot(data)
_ = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
_ = ax.set_ylabel("exclusion diameter in milimeters")
```

This version is using the mean as a point, and standard deviation as an error bar directly and is not calculating a box height.

```python
matplotlib.style.use("ggplot")

f = pyplot.figure(figsize=(6,4))
ax = f.add_subplot(1,1,1)

_ = ax.errorbar(data.columns, data.mean(), yerr=data.std(), 
                linestyle="", 
                color="purple", 
                marker=".", 
                ecolor="r",
                capsize=5,
               )
_ = ax.set_ylabel("Average zone of inhibition (mm)")
_ = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
```
